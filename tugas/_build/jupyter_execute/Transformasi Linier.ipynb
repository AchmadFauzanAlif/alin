# Transformasi Linier

## Definisi Transformasi Linier
Transformasi linear yang dimaksud adalah perpindahan
dari satu ruang yang biasa dinamakan dengan domain ke ruang lain yang
dinamakan kodomain. Salah satu pembahasan dalam perkuliahan aljabar
adalah mengenai transformasi linear yaitu suatu fungsi yang dapat
memetakan suatu ruang vektor ke ruang vektor yang lain, sehingga operasi
standar pada ruang vektor (penjumlahan dan perkalian dengan skalar) tetap
berlaku.

## Sifat Transformasi Linier
Sifat sifat dari transformasi linier adalah sebagai berikut: 


1. Pelestarian Jumlah: 
Transformasi linier mempertahankan jumlah dua vektor. Artinya, jika kita mempunyai dua vektor $u$ dan $v$, transformasi linier $T$ memenuhi $$ T(u+v) = T(u) + T(v) $$
Pembuktiannya adalah sebagai berikut ini:

$$
A = \begin{pmatrix}
1 & 2 \\ 
3 & 4
\end{pmatrix}
$$

Ambil dua vektor $u$ dan $v$

$$
u = \begin{pmatrix}
1\\
2
\end{pmatrix},

v = \begin{pmatrix}
2\\
3
\end{pmatrix}

$$

pertama, kita hitung $t(u)$ dan $t(v)$

$$

t(u) = Au = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} 
\begin{pmatrix} 1 \\ 1 \end{pmatrix} =
\begin{pmatrix} 1 &  1 & + & 2  & 1 \\ 3 & 1 & + & 4 & 1\end{pmatrix} =
\begin{pmatrix} 3 \\ 7 \end{pmatrix}

$$

$$

t(u) = Au = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} 
\begin{pmatrix} 2 \\ 3 \end{pmatrix} =
\begin{pmatrix} 1 &  2 & + & 2  & 3 \\ 3 & 2 & + & 4 & 3\end{pmatrix} =
\begin{pmatrix} 8 \\ 18 \end{pmatrix}

$$

$$

t(u) + t(v) =
\begin{pmatrix} 3 \\ 7 \end{pmatrix} +
\begin{pmatrix} 8 \\ 18 \end{pmatrix} =
\begin{pmatrix} 11 \\ 25 \end{pmatrix}

$$

Selanjutnya kita coba hitung $t(u+v)$:

$$ 

u + v = 
\begin{pmatrix} 1 \\ 1 \end{pmatrix} +
\begin{pmatrix} 2 \\ 3 \end{pmatrix} = 
\begin{pmatrix} 3 \\ 4 \end{pmatrix}

$$

$$

t(u + v) = A(u + v) = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}
\begin{pmatrix} 3 \\ 4 \end{pmatrix} =
\begin{pmatrix} 1 & 3 & + & 2 & 4 \\ 3 & 3 & + & 4 & 4 \end{pmatrix} = 
\begin{pmatrix} 11 \\ 25 \end{pmatrix}

$$





2. Pelestarian Perkalian dengan Skalar:
Transformasi linier juga mempertahankan perkalian suatu vektor dengan skalar. artinya jika mempunyai vektor $u$ dan skalar $c$, transformasi linier $T$ memenuhi 
$ T(\alpha u) = \alpha T(u) $

Pembuktian dari rumus ini adalah sebagai berikut ini: 

$$ A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} $$

Ambil vektor $u$

$$ u = \begin{pmatrix} 1 \\ 1 \end{pmatrix} $$

dan ambil skalar $a$:

$ a = 3 $ 

pertama, kita hitung $ T(\alpha u) $ :

a. kalikan vektor $u$ dengan skalar $\alpha$:
$$ 
\alpha u = 3 \begin{pmatrix} 1 \\ 1 \end{pmatrix} = 
\begin{pmatrix} 3 \\ 3 \end{pmatrix} 
$$

b. Terapkan transformasi $T$ pada vektor $\alpha u$:

$$ 
T(\alpha u) = a(\alpha u) = 
\begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} 
\begin{pmatrix} 3 \\ 3 \end{pmatrix} = 
\begin{pmatrix} 1 & 3 & + & 2 & 3 \\ 3 & 3 & + & 4 & 3 \end{pmatrix} = 
\begin{pmatrix} 9 \\ 21 \end{pmatrix}
$$

Kedua, kita hitung $ \alpha T(u): $

a. Terapkan transformasi $T$ pada vektor $u$:

$$
T(u) = Au = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} 
\begin{pmatrix} 1 \\ 1 \end{pmatrix} = 
\begin{pmatrix} 3 \\ 7 \end{pmatrix}
$$

b. Kalikan hasil transformasi $T(u)$ dengan skalar $\alpha$

$$

aT(u) = 3 \begin{pmatrix} 3 \\ 7 \end{pmatrix} = \begin{pmatrix} 9 \\ 21 \end{pmatrix}

$$

## Penerapan dari Transformasi Linier

Soal: 

$$

T \begin{pmatrix} 2 \\ 0 \end{pmatrix} = \begin{pmatrix} -2 \\ 0 \end{pmatrix} 
$$
$$
T \begin{pmatrix} 2 \\ 2 \end{pmatrix} = \begin{pmatrix} -2 \\ 2 \end{pmatrix}

$$

Mari kita mulai dengan menuliskan bentuk umum matrix A yang memetakan vektor - vektor standart: 

$$
A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}
$$

kita perlu menemukan $a,b,c,d$ sehingga syarat-syarat di atas terpenuhi

1. Menggunakan informasi pertama

$$
A \begin{pmatrix} 2 \\ 0 \end{pmatrix} = \begin{pmatrix} -2 \\ 0 \end{pmatrix}
$$

ini memberikan kita dua persamaan:
$$
\begin{pmatrix} a & b \\ c & d \end{pmatrix} \begin{pmatrix} 2 \\ 0 \end{pmatrix} = \begin{pmatrix} 2a \\ 2c \end{pmatrix} = \begin{pmatrix} -2 \\ 0 \end{pmatrix}
$$

Dari sini, kita mendapatkan: 

a. $2a = -2 \Longrightarrow \alpha = -1 \\$ 
 
b. $2c = 0 \Longrightarrow c =0 $

2. Menggunakan informasi kedua:
$$

A \begin{pmatrix} 2 \\ 2 \end{pmatrix} = 
\begin{pmatrix} -2 \\ 2 \end{pmatrix}
$$

ini memberikan kita dua persamaan: 
$
\begin{pmatrix} a & b \\ c & d \end{pmatrix} \begin{pmatrix} 2 \\ 2 \end{pmatrix} = \begin{pmatrix} 2 \\ 2 \end{pmatrix} = \begin{pmatrix} 2a + 2b \\ 2c + 2d \end{pmatrix} = \begin{pmatrix} -2 \\ 2 \end{pmatrix}
$

dari sini, kita mendapatkan: 


a. $ 2a + 2b = -2 \\$
b. $ 2c + 2d = 2$

dengan $a = -1 $ dan $ c = 0 $, subsitusi ke dalam persamaan ini menghasilkan

1. $ 2(-1) + 2b = -2 \Longrightarrow -2 + 2b = -2 \Longrightarrow 2b = 0 \Longrightarrow b = 0 \\$
2. $ 2(0) + 2d = 2 \Longrightarrow 2d = 2 \Longrightarrow d = 1 $

### Matrix Transformasi

Setelah menentukan nilai-nilai dari $a,b,c,d$ kita mendapatkan: 

$
A \begin{pmatrix} -1 & 0 \\ 0 & 1 \end{pmatrix}
$

## Pengertian dari Refleksi pada transformasi linier

Refleksi dalam transformasi linier adalah sebuah operasi yang mengubah sifat vektor dalam ruang vektor. Dalam konteks transformasi linier, refleksi dapat diartikan sebagai perubahan orientasi vektor yang terjadi ketika vektor tersebut dipantulkan terhadap suatu bidan atau sumbu. 

Misalnya, jika kita memiliki vektor $v$ dan kita ingin melakukan reflesi terhadap sumbu $x$, maka hasilnya akan menjadi -$v$. Artinya, vektor asli $v$ akan dipantulkan terhadap sumbu $x$ sehingga menjadi vektor baru -$v$ yang memiliki orientasi yang berlawanan dengan vektor asli.
